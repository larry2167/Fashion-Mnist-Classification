## Fashion MNIST Dataset comperasion Basemodel Vs AlexNet

```python
 # importing all libraries 
import itertools
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout,Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.metrics import SparseCategoricalAccuracy
```

```python
# setup GPU
if True:
    print("GPU device is set:", tf.config.list_physical_devices('GPU'))
else:
    print("No GPU devices found.")
```

## Load Dataset and Data preprocessing

```python
# load the Dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
train_images = train_images / 255.0
test_images = test_images / 255.0
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
num_classes = 10
```

```python
import random
plt.figure(figsize=(6, 3))
for i in range(6):
  ax = plt.subplot(2, 3, i + 1)
  rand_index = random.choice(range(len(train_images)))
  plt.imshow(train_images[rand_index])
  plt.title(class_names[train_labels[rand_index]])
  plt.axis(False)
```

## Creating NN (BaseModel and AlexNet)

```python
class BaseModelFashionMNIST(tf.keras.Model):
    def __init__(self, num_classes:int):
        super(BaseModelFashionMNIST, self).__init__()
        self.model = Sequential([
                                  Flatten(),
                                  Dense(100, activation='relu'),
                                  Dropout(0.2),
                                  Dense(num_classes, activation='softmax')
                                ])

    def call(self, inputs):
        return self.model(inputs)

model_0 = BaseModelFashionMNIST(num_classes)
```

```python
class AlexNetFashionMNIST(tf.keras.Model):
    def __init__(self, num_classes):
        super(AlexNetFashionMNIST, self).__init__()
        self.conv_layers = Sequential([
                                        Conv2D(96, (3, 3), strides=(1, 1), activation='relu', input_shape=(28, 28, 1)),
                                        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
                                        Conv2D(256, (5, 5), padding='same', activation='relu'),
                                        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
                                        Conv2D(384, (3, 3), padding='same', activation='relu'),
                                        Conv2D(384, (3, 3), padding='same', activation='relu'),
                                        Conv2D(256, (3, 3), padding='same', activation='relu'),
                                        MaxPooling2D(pool_size=(2, 2), strides=(2, 2))
                                      ])

        self.fc_layers = Sequential([
                                      Flatten(),
                                      Dense(4096, activation='relu'),
                                      Dropout(0.5),
                                      Dense(4096, activation='relu'),
                                      Dropout(0.5),
                                      Dense(num_classes, activation='softmax')
                                    ])

    def call(self, inputs):
        x = self.conv_layers(inputs)
        x = self.fc_layers(x)
        return x

model_1 = AlexNetFashionMNIST(num_classes)
```

```python

def compile_and_fit(model, train_images, train_labels, test_images,
                    test_labels, epochs=10, batch_size=32):
    optimizer = Adam(learning_rate=0.001)
    loss = SparseCategoricalCrossentropy()
    metrics = SparseCategoricalAccuracy()
    model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                        validation_data=(test_images, test_labels))
    return history

def plot_loss_accuracy(loss_df, common_title):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    ax1.plot(loss_df['loss'], color='red', label='Train Loss')
    ax1.plot(loss_df['val_loss'], color='blue', label='Validation Loss')

    ax1.set_title('Loss over epochs')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')

    ax2.plot(loss_df['sparse_categorical_accuracy'], color='red', label='Train Accuracy')
    ax2.plot(loss_df['val_sparse_categorical_accuracy'], color='blue', label='Validation Accuracy')

    ax2.set_title('Accuracy over epochs')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax1.legend()
    fig.suptitle(common_title, fontweight='bold', fontsize=14, fontfamily='serif')
```

```python
basemodel_history = compile_and_fit(model_0, train_images, train_labels, test_images,test_labels, epochs=10, batch_size=32)
basemodel_df = pd.DataFrame(basemodel_history.history)
```

##plot the epochs history



```python
plot_loss_accuracy(basemodel_df, common_title='BaseModelFashionMNIST')
```

```python
alexnet_history = compile_and_fit(model_1, train_images, train_labels, test_images,test_labels, epochs=10, batch_size=32)
alexnet_df = pd.DataFrame(alexnet_history.history)
```

```python
plot_loss_accuracy(alexnet_df, common_title='AlexNetFashionMNIST')
```

```python
def plot_random_image(model, images, true_labels, classes):
    plt.figure(figsize=(7, 4))
    for i in range(4):
        ax = plt.subplot(2, 2, i + 1)
        rand_index = random.choice(range(len(test_images)))
        target_image = images[rand_index]
        pred_probs = model.predict(target_image.reshape(1, 28, 28), verbose=0)
        pred_label = classes[pred_probs.argmax()]
        true_label = classes[true_labels[rand_index]]

        plt.imshow(target_image)

        if pred_label == true_label:
            color = "green"
        else:
            color = "red"
        
        plt.title("Pred: {} {:2.0f}% (True: {})".format(pred_label,
                                                        100 * tf.reduce_max(pred_probs),
                                                        true_label),
                  color=color, fontsize=10)
        
    plt.tight_layout()
plot_random_image(model_1, test_images, test_labels, class_names)
plt.show()
```

```python
y_preds = model_1.predict(test_images)
y_preds = y_preds.argmax(axis=1)
y_preds
```

## ploting confusion metrix

```python
def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15):
    cm = confusion_matrix(y_true, y_pred)
    cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis]
    n_classes = cm.shape[0]

    fig, ax = plt.subplots(figsize=figsize)
    cax = ax.matshow(cm, cmap=plt.cm.Blues)
    fig.colorbar(cax)

    if classes:
        labels = classes
    else:
        labels = np.arange(cm.shape[0])

    ax.set(title="Confusion Matrix",
           xlabel="Predicted label",
           ylabel="True label",
           xticks=np.arange(n_classes),
           yticks=np.arange(n_classes),
           xticklabels=labels,
           yticklabels=labels)

    ax.xaxis.set_label_position("bottom")
    ax.xaxis.tick_bottom()

    threshold = (cm.max() + cm.min()) / 2.

    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j]*100:.1f}%)",
                 horizontalalignment="center",
                 color="white" if cm[i, j] > threshold else "black",
                 size=text_size)

# Usage example
make_confusion_matrix(y_true=test_labels, y_pred=y_preds, classes=class_names, figsize=(10,6), text_size=6)
plt.show()
```
